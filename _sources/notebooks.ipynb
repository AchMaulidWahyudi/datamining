```python
import numpy as np
import pandas as pd
```

```python
dataset_url="https://raw.githubusercontent.com/AchMaulidWahyudi/datamining/main/ds_salaries.csv"
data=pd.read_csv(dataset_url, index_col=0)
```

```python
data
```

```python
def diss(x,y):
  a = 2
  b = 0
  for col in ['salary','salary_in_usd']:
    if data[col][x] == data[col][y]:
      b+=1
  return (a-b)/a
```

**Menghitung Perbedaan**

```python
print(diss(1,1))
print(diss(1,2))
print(diss(3,3))
```
